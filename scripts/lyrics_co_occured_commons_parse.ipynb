```python
import pandas as pd

url = "https://raw.githubusercontent.com/JunHCha/K-POP-Lyrics-1964-2020/main/data/lyrics_cooccured/lyric_cooccured_common_words_1995_2019.csv"
words = pd.read_csv(url, encoding="utf-8").iloc[:, 1:]
words
```

```python
word_fragments_by_year = [
    words.loc[words["year"].isin(period)]
    for period in [
        list(range(1995, 2000)),
        list(range(2000, 2005)),
        list(range(2005, 2010)),
        list(range(2010, 2015)),
        list(range(2015, 2020)),
    ]
]
for word in word_fragments_by_year:
    file_name = f"lyric_cooccured_commons_{int(word['year'].min())}_{int(word['year'].max())}"
    print(file_name)
    word.reset_index(drop=True).to_csv(
        f"../data/lyrics_cooccured/{file_name}.csv", encoding="utf-8-sig"
    )
```
