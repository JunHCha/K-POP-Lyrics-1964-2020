```python
import pandas as pd
import datetime as dt

url = "https://raw.githubusercontent.com/JunHCha/K-POP-Lyrics-1964-2020/main/data/lyrics_cooccured/cooccured_words_by_line.csv"
words = pd.read_csv(url, encoding="utf-8").iloc[:, 1:]
words
```

```python
periods = [
    list(range(1986, 1992)),
    list(range(1992, 1998)),
    list(range(2015, 2021)),
]
```

```python
for period in periods:
    words_in_period = words[words["year"].isin(period)]
    edges = (
        words_in_period.groupby(["noun", "predicate"]).size().reset_index(name="weight")
    )
    edges = edges.loc[
        edges["weight"] > edges.quantile(0.95).weight
    ]
    edges.index.names = ["ids"]
    edges.columns = ["Source", "Target", "Weight"]
    edges.to_csv(
        f"../data/lyrics_cooccured/cooccured_edges_in_period_{period[0]}to{period[-1]}.csv",
        encoding="utf-8",
        index=False,
    )
```
